```python
import pandas as pd 
```

```python
#Loading the dataset
df = pd.read_csv('https://raw.githubusercontent.com/amirtds/kaggle-netflix-tv-shows-and-movies/refs/heads/main/titles.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
#Checking for null values
```

```python
#Cheking for any duplicates
df.duplicated().sum()
```

```python
#Dropped all null values
df = df.dropna()
```

```python
#Making sure that all null values are gone
df.isna().sum()
```

```python
#How many shows ans movies exist?
type_counts = df['type'].value_counts()
type_counts
```

```python
#What is the most common genre?
most_common_genre = df['genres'].value_counts().idxmax()
most_common_genre
```

```python
#In what year were most shows and movies released?
most_releases_year = df['release_year'].value_counts().idxmax()
most_releases_year
```

```python
#Which country has the most releases?
most_content_country = df['production_countries'].value_counts().idxmax()
most_content_country
```

```python
#Create a new column that displays how long it has been since a particular content was released
df['content_age']= 2025 - df['release_year']
```

```python
#Confirming that the column has been created
df.head(2)
```

```python
#We're finding out the ratio of movies to series released
percentage_split = df['type'].value_counts(normalize=True)*100
percentage_split
```

```python
#Filtering out content thatb have a high imdb score
high_quality_scores = df[df['imdb_score']> 7.5].iloc[:]
high_quality_scores
```

```python
#Of all the shows that have a high imdb score, which ones are the best five?
top_5_high_quality =df[df['imdb_score']> 7.5].sort_values(by='imdb_score', ascending=False).iloc[0:4,[1,2,11]]
top_5_high_quality
```

```python
#Finding out which content got high traffic but still scored very poorly
high_traffic_lower_rates = df[(df['imdb_votes'] >= 50000) & (df['imdb_score'] < 6.5)]
high_traffic_lower_rates
```

```python
top_five = df.groupby('production_countries')['type'].value_counts(['type']).head()
top_five
```
